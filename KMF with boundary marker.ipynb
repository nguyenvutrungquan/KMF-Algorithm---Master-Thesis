```python
"""
A numerical test for the Cauchy problem for the Poisson equation 
with relaxed Kozlov-Maz'ya's algorithm 2, by using FEniCS project.
-Delta(u) = f in \Omega,
\frac{partial u}{\partial \nu} = g_1 on \Gamma_1,
u = v_1 on \Gamma_1.

Reference: Solving PDEs in Python  The FEniCS Tutorial I 
by Hans Petter Langtangen and Anders Logg.
Link: https://fenicsproject.org/tutorial/
"""
#Libraries
from fenics import *
from mshr import *
import matplotlib.pyplot as plt
import math as m
import numpy as np
from time import time


#Define Function Space and Mesh,
domain_vertices = Rectangle(Point(0.0,0.0), Point(1.0,1.0))
mesh= generate_mesh(domain_vertices, 64)
V = FunctionSpace(mesh, 'P', 1) #Lagrange element of degree 1.

#Define datum of the Cauchy problem.
v_1 = Expression('1+3*x[0]*x[0]+2*x[1]*x[1]', degree=10)
g_1 = Expression('4*x[1]', degree=10)
f = Constant(-6.0)
epsilon = 1E-6
tol = 1E-14 #Use for defining the boundaries, since the exact comparison == is not good in programming pratice.
dGamma_0 = FacetNormal(mesh) #Define unit vector normal to the entire boundary Gamma.
u_exact = Expression('1+3*x[0]*x[0]+2*x[1]*x[1]', degree=10)

#Define boundaries. In this test we do not need to use it.
boundary_markers = MeshFunction("size_t", mesh, 1)
Gamma_0 = CompiledSubDomain('on_boundary && (near(x[0],0,tol) || near(x[0],1,tol))', tol=tol)
Gamma_0.mark(boundary_markers, 0)
Gamma_1 = CompiledSubDomain('on_boundary && (near(x[1],0,tol) || near(x[1],1,tol))', tol=tol)
Gamma_1.mark(boundary_markers, 1)

#Define the PDE solvers
def solver1(u_D0, u_N1):
    #Define boundary condition
    bc1 = DirichletBC(V,u_D0, Gamma_0)
    #Define variational problem
    usolver1 = TrialFunction(V)
    v = TestFunction(V)
    a_1 = dot(grad(usolver1), grad(v))*dx
    #L_1 = f*v*dx - u_N1*v*ds
    #or 
    L_1 = f*v*dx - u_N1*v*ds(1) #if you want to use the boundary marker above.
    
    #Compute solution
    usolver1 = Function(V)
    solve(a_1 == L_1, usolver1, bc1)
    return usolver1

def solver2(u_N0, u_D1):
    #Define boundary condition
    bc2 = DirichletBC(V,u_D1, Gamma_1)
    #Define variational problem
    usolver2 = TrialFunction(V)
    v = TestFunction(V)
    a_2 = dot(grad(usolver2), grad(v))*dx
    #_2 = f*v*dx - u_N0*v*ds
    #or 
    L_2 = f*v*dx - u_N0*v*ds(0) #if you want to use the boundary marker above.
    
    #Compute solution
    usolver2 = Function(V)
    solve(a_2 == L_2, usolver2, bc2)
    return usolver2

#Perform Kozlov-Maz'ya Algorithm
# theta = 1
# Define the stopping criterion
def error(x,y,z,t):
    return abs(errornorm(x,y,'L2') - errornorm(z,t,'L2'))
#Define the number of iterations as a function of relaxation parameter.
def iteration(theta): 
#Step 1. Specify an initial guess a_0
    a_0 = Expression('x[0]', degree = 10)
#Step 2. Find u_0 and obtain b_0, find u_1 and obtain a_1
    u_0 = solver1(a_0, g_1)
    b_0 = dot(grad(u_0),dGamma_0)
    u_1 = solver2(b_0, v_1)
    a_1 = theta*u_1 + (1-theta)*a_0
    num_iterations = 1

    u_2 = solver1(a_1, g_1)
    b_1 = dot(grad(u_2),dGamma_0)
    u_3 = solver2(b_1, v_1)
    a_0 = theta*u_3 + (1-theta)*a_1
    num_iterations = 2

#Step 3.
    if error(u_0, u_1, u_2, u_3) < epsilon:
        print('Stop at iteration ', num_iterations)
    else:
        while error(u_0, u_1, u_2, u_3) >= epsilon:
            u_0 = solver1(a_0, g_1)
            b_0 = dot(grad(u_0),dGamma_0)
            u_1 = solver2(b_0, v_1)
            """
            Since the test function v vanishes on the Dirichlet boundary Gamma_1, 
            I use the solver2 function with the normal derivative b_0, which is defined on the entire boudary Gamma.
            See the reference above, page 85 for explanation.
            """
            a_1 = theta*u_1 + (1-theta)*a_0
            num_iterations += 1
    
            u_2 = solver1(a_1, g_1)
            b_1 = dot(grad(u_2),dGamma_0) 
            u_3 = solver2(b_1, v_1) 
            a_0 = theta*u_3 + (1-theta)*a_1
            num_iterations += 1
            print(error(u_0, u_1, u_2, u_3))
            if num_iterations >300:
                break #To limit the number of iterations in case the method does not converge.
            if error(u_0, u_1, u_2, u_3) >= 4:
                num_iterations = 300
                break #Another condition to limit the number of iterations in case the method does not converge.
    return num_iterations

#Plot function with theta from 0.1 to 2.0, I take 20 values.
theta = []
iters = []
for i in range(20):
    theta.append(0.1 +i/10)
    iters.append(iteration(theta[i]))
plt.figure()
plt.plot(theta, iters,'b')
plt.title('Variation of the Number of iterations as a function of Relaxation Parameter')
plt.xlabel('Relaxation parameter')
plt.ylabel('Number of iterations')
plt.grid(True)
plt.savefig('relaxation algorithm 2.png', dpi=300, bbox_inches='tight')
plt.savefig('relaxation algorithm 2.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
